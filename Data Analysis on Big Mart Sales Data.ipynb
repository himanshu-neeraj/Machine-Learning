# **Data Analysis on Big Mart Sales Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_BM = pd.read_csv(r'/Users/himanshuneeraj/Documents/AAFT Classes /Semester -2 /Machine Learning /Haadiya project/big_mart_sales.csv')
data_BM
```

```python
# display first 5 rows
data_BM.head()
```

```python
# check shape of dataset
data_BM.shape
```

```python
# dataset information
data_BM.info()
```

```python
# statistical summary
data_BM.describe()
```

```python
# check missing values
data_BM.isnull().sum()
```

```python
# countplot for outlet size
sns.countplot(x='Outlet_Size', data=data_BM)

plt.title("Count of Outlet Sizes")

plt.show()
```

```python
# Observation
# The countplot shows that medium-sized outlets are more common in the dataset compared to small and high-sized outlets.
```

```python
# histogram of item outlet sales
plt.hist(data_BM['Item_Outlet_Sales'], bins=20)

plt.xlabel("Item Outlet Sales")
plt.ylabel("Frequency")
plt.title("Distribution of Item Outlet Sales")

plt.show()
```

```python
# Observation
# The histogram shows that most Item_Outlet_Sales values are concentrated in the lower to middle range,
# while fewer items have very high sales.
```

```python
# distribution plot
sns.distplot(data_BM['Item_MRP'])

plt.title("Distribution of Item MRP")

plt.show()
```

```python
# Observation
# The distribution plot shows that Item_MRP values are spread across different price ranges, with most items concentrated in the
# middle range.
```

```python
# sales by outlet size
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

# sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

# set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'orange', 'magenta'])

# add labels and title
plt.xlabel("Outlet Size")
plt.ylabel("Sales")
plt.title("Mean Sales for each outlet type")

plt.show()
```

```python
# Observation
# The plot shows the average sales for different outlet sizes. Medium-sized outlets appear to generate higher sales
#compared to small and high-sized outlets.
```

## **Scatter plot of Item Weight vs Item Visibility.**

```python
plt.scatter(data_BM['Item_Weight'], data_BM['Item_Visibility'])

plt.xlabel("Item Weight")
plt.ylabel("Item Visibility")
plt.title("Item Weight vs Item Visibility")

plt.show()
```

```python
# Observation
# The scatter plot indicates that there is no strong relationship between item weight and item visibility.
# The data points are scattered randomly.
```

## **Box plot of Item Outlet Sales**

```python
sns.boxplot(y=data_BM['Item_Outlet_Sales'])

plt.title("Box Plot of Item Outlet Sales")

plt.show()
```

```python
# Observation
# The box plot shows the distribution of outlet sales and highlights the presence of outliers in the sales data.
# Most sales values are concentrated within a specific range, while a few items have exceptionally high sales.
```

```python
# correlation matrix
corr = data_BM.corr(numeric_only=True)

plt.figure(figsize=(10,6))

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```

```python
# Observation

# The heatmap shows that Item_MRP has the strongest positive correlation with Item_Outlet_Sales, while most other variables have
# weak correlations.
```

```python
# outlet establishment year count
sns.countplot(x='Outlet_Establishment_Year', data=data_BM)

plt.title("Outlet Establishment Year Count")

plt.show()
```

```python
# Observation
# The plot shows the number of outlets established in different years, with some years having significantly more outlets than others.
```

```python
# average sales by outlet type
avg_sales = data_BM.groupby('Outlet_Type')['Item_Outlet_Sales'].mean()

avg_sales.plot(kind='bar')

plt.xlabel("Outlet Type")
plt.ylabel("Average Sales")
plt.title("Average Sales by Outlet Type")

plt.show()
```

```python
# Observation
# The plot shows that supermarket outlets have higher average sales compared to grocery stores.
```

```python
# item fat content count
sns.countplot(x='Item_Fat_Content', data=data_BM)

plt.title("Item Fat Content Count")

plt.show()
```

```python
# Observation
# The plot shows that Low Fat and Regular items are the most common categories in the dataset.
```

```python
# pairplot of numerical columns
sns.pairplot(data_BM[['Item_Weight',
                      'Item_Visibility',
                      'Item_MRP',
                      'Item_Outlet_Sales']])

plt.show()
```

```python
# Observation
# The pairplot shows the relationship between numerical variables such as Item_Weight, Item_Visibility, Item_MRP, and Item_Outlet_Sales.
# A positive relationship can be observed between Item_MRP and Item_Outlet_Sales, indicating that products with higher MRP tend to
# generate higher sales. Most other variable pairs show weak or scattered relationships with no strong correlation.
```
